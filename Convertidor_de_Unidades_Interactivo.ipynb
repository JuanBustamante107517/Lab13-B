```python
# Definimos las tasas de conversión usando diccionarios anidados.
# Esto nos permite organizar las conversiones por tipo (Longitud, Temperatura).

# Diccionario para conversiones de longitud.
# La clave es la unidad de origen, y el valor es otro diccionario
# que mapea la unidad de destino a su factor de conversión.
CONVERSIONES_LONGITUD = {
    "metros": {"kilometros": 0.001, "centimetros": 100, "millas": 0.000621371, "pies": 3.28084},
    "kilometros": {"metros": 1000, "centimetros": 100000, "millas": 0.621371, "pies": 3280.84},
    "centimetros": {"metros": 0.01, "kilometros": 0.00001, "millas": 0.0000062137, "pies": 0.0328084},
    "millas": {"metros": 1609.34, "kilometros": 1.60934, "centimetros": 160934, "pies": 5280},
    "pies": {"metros": 0.3048, "kilometros": 0.0003048, "centimetros": 30.48, "millas": 0.000189394}
}

# Diccionario para conversiones de temperatura (estas son más complejas, se harán con funciones).
# Aquí solo indicamos las unidades posibles para cada tipo de temperatura.
CONVERSIONES_TEMPERATURA = {
    "celsius": ["fahrenheit", "kelvin"],
    "fahrenheit": ["celsius", "kelvin"],
    "kelvin": ["celsius", "fahrenheit"]
}

# --- Funciones de Conversión de Temperatura (Paso 2) ---
# Función para convertir Celsius a Fahrenheit.
def celsius_a_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Función para convertir Fahrenheit a Celsius.
def fahrenheit_a_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9

# Función para convertir Celsius a Kelvin.
def celsius_a_kelvin(celsius):
    return celsius + 273.15

# Función para convertir Kelvin a Celsius.
def kelvin_a_celsius(kelvin):
    return kelvin - 273.15

# Función para convertir Fahrenheit a Kelvin (pasando por Celsius).
def fahrenheit_a_kelvin(fahrenheit):
    celsius = fahrenheit_a_celsius(fahrenheit) # Primero convertimos a Celsius.
    return celsius_a_kelvin(celsius) # Luego de Celsius a Kelvin.

# Función para convertir Kelvin a Fahrenheit (pasando por Celsius).
def kelvin_a_fahrenheit(kelvin): # <--- Corrección de 'farenheit' a 'fahrenheit'
    celsius = kelvin_a_celsius(kelvin) # Primero convertimos a Celsius.
    return celsius_a_fahrenheit(celsius) # Luego de Celsius a Fahrenheit.


# --- Bucle principal del programa (Paso 1 y control principal) ---
while True:
    print("\n--- Convertidor de Unidades ---")
    print("Selecciona un tipo de conversión:")
    print("1. Longitud")
    print("2. Temperatura")
    print("3. Salir")

    tipo_conversion = input("Ingresa el número de la opción deseada: ")

    # --- Lógica de opciones (Paso 3, 4 y 5) ---
    # ¡Estas líneas deben estar indentadas para estar dentro del bucle while!
    if tipo_conversion == "1":  # Opción de Longitud
        print("\n--- Conversión de Longitud ---")
        print("Unidades disponibles: metros, kilometros, centimetros, millas, pies")
        try:
            unidad_origen = input("Ingresa la unidad de origen: ").lower().strip()
            unidad_destino = input("Ingresa la unidad de destino: ").lower().strip()
            valor = float(input("Ingresar el valor a convertir: "))

            if unidad_origen in CONVERSIONES_LONGITUD and unidad_destino in CONVERSIONES_LONGITUD[unidad_origen]:
                factor = CONVERSIONES_LONGITUD[unidad_origen][unidad_destino]
                resultado = valor * factor
                # Añadido espacio entre valor y unidad_origen
                print(f"{valor} {unidad_origen} son {resultado:.4f} {unidad_destino}")
            else:
                print("Unidades de longitud no válidas o conversión no soportada. Intenta de nuevo.")
        except ValueError: # <--- Corrección: ValuError a ValueError
            print("Valor inválido. Por favor, ingresa un número.")

    elif tipo_conversion == "2": # Opción de Temperatura
        print("\n--- Conversión de Temperatura ---")
        print("Unidades disponibles: celsius, fahrenheit, kelvin")
        # El try debe estar indentado para pertenecer al elif
        try:
            unidad_origen = input("Ingresa la unidad de origen: ").lower().strip()
            unidad_destino = input("Ingresa la unidad de destino: ").lower().strip()
            valor = float(input("Ingresa el valor a convertir: "))

            resultado = None # Iniciamos resultado como None

            # Lógica para llamar a la función de conversión correcta.
            if unidad_origen == "celsius":
                if unidad_destino == "fahrenheit":
                    resultado = celsius_a_fahrenheit(valor)
                elif unidad_destino == "kelvin": # <--- Corrección: Añadido el ':'
                    resultado = celsius_a_kelvin(valor)
            elif unidad_origen == "fahrenheit":
                if unidad_destino == "celsius":
                    resultado = fahrenheit_a_celsius(valor)
                elif unidad_destino == "kelvin":
                    resultado = fahrenheit_a_kelvin(valor)
            elif unidad_origen == "kelvin":
                if unidad_destino == "celsius":
                    resultado = kelvin_a_celsius(valor)
                elif unidad_destino == "fahrenheit":
                    resultado = kelvin_a_fahrenheit(valor) # <--- Corrección: a kelvin_a_fahrenheit

            if resultado is not None:  # Si se encontró una conversión válida.
                print(f"{valor} {unidad_origen} son {resultado:.2f} {unidad_destino}")  # 2 decimales para temperatura
            else:
                print("Unidades de temperatura no válidas o conversión no soportada. Intenta de nuevo.")
        except ValueError: # <--- Corrección: ValuError a ValueError
            print("Valor inválido. Por favor, ingresa un número.")

    elif tipo_conversion == "3":
        print("¡Gracias por usar el Convertidor de Unidades! ¡Hasta luego!")
        break # Sale del bucle while True

    else: # Esta opción debe estar indentada para pertenecer al while True
        print("Opción no válida. Por favor, elige 1, 2 o 3.")
```
