##Inicializar la lista de tareas y mostrar el menú principal

```python
# Iniciamos una lista vacía para almacenar las tareas.
# Una lista es una colección ordenada y modificable de elementos.
tareas = []

# Iniciamos un bucle 'while True'.
# Este crea un bucle infinito, lo que significa que el menú se mostrará repetidamente
# hasta que el usuario decida explícitamente salir (usando 'break').
while True:
    # Mostramos el menú de opciones al usuario.
    print("\n--- Gestor de Tareas ---")
    print("1. Agregar tarea")
    print("2. Ver tareas")
    print("3. Eliminar tarea")
    print("4. Salir")

    # Pedimos al usuario que elija una opción.
    opcion = input("Elige una opción: ")

    # Usamos un 'if-elif-else' para manejar las diferentes opciones del menú.
    # ¡Estas líneas deben estar indentadas para estar dentro del bucle while!
    if opcion == '1':
        # Si el usuario elige '1', le pedimos la descripción de la nueva tarea.
        nueva_tarea = input("Ingresa la nueva tarea: ")
        # Usamos el método '.append()' para añadir la nueva tarea al final de la lista 'tareas'.
        tareas.append(nueva_tarea) # <--- ¡Corrección aquí! Faltaba el .append()
        print(f"Tarea '{nueva_tarea}' agregada.")

    elif opcion == '2':
        # Si el usuario elige '2', mostramos todas las tareas.
        print("\n--- Tus Tareas ---")
        if not tareas:
            print("No hay tareas pendientes.")
        else:
            # Usamos un bucle 'for' para recorrer cada tarea en la lista.
            # 'enumerate' nos da tanto el índice (posición) como el valor de cada elemento.
            # Empezamos el conteo en 1 para que las tareas se muestren como 1., 2., etc.
            for i, tarea in enumerate(tareas, 1):
                print(f"{i}. {tarea}")

    elif opcion == '3':
        # Si el usuario elige '3', primero mostramos las tareas para que elija cuál eliminar.
        if not tareas:
            print("No hay tareas para eliminar.")
        else:
            print("\n--- Tareas Actuales ---")
            for i, tarea in enumerate(tareas, 1):
                print(f"{i}. {tarea}")

            try:
                # Pedimos al usuario el número de la tarea a eliminar.
                num_tarea = int(input("Ingresa el número de la tarea a eliminar: "))
                # Convertimos el número a un índice de lista (restamos 1 porque las listas son base 0).
                indice_tarea = num_tarea - 1

                # Verificamos si el índice está dentro del rango válido de la lista.
                if 0 <= indice_tarea < len(tareas): # <--- ¡Corrección aquí! Era len(tarea) y debe ser len(tareas)
                    # Usamos '.pop()' para eliminar la tarea en el índice especificado.
                    # '.pop()' también devuelve el elemento eliminado, lo cual es útil.
                    tarea_eliminada = tareas.pop(indice_tarea)
                    print(f"Tarea '{tarea_eliminada}' eliminada.")
                else:
                    print("Número de tarea no válido.")
            except ValueError:
                print("Entrada no válida. Por favor, ingresa un número.")

    elif opcion == '4':
        # Si el usuario elige '4', mostramos un mensaje de despedida y salimos del bucle.
        print("¡Gracias por usar el Gestor de Tareas! ¡Hasta luego!")
        break # La instrucción 'break' saca el programa del bucle 'while True'.
    else:
        # Si la opción ingresada no es ninguna de las válidas (1, 2, 3, 4), mostramos un error.
        print("Opción no válida. Por favor, elige un número del 1 al 4.") # <--- ¡Corrección aquí! Mensaje incompleto.
```
